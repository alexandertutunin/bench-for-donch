```python
import json
import pandas as pd

with open("./performance.json", "r") as f:
    data = json.load(f)
```

```python
demo = pd.DataFrame.from_records(data)
demo.set_index("device", drop=True, inplace=True)
demo["Score"] = demo["ep/s"] / demo.loc["NVIDIA GeForce GTX 1050"]["ep/s"].item()
demo["hrs for 300 epochs"] = demo["s/ep"] * 300 /60 /60
# demo["Score"] = demo["ep/s"] / demo["ep/s"].min()
demo
```

```python
demo = demo[["Score", "ep/s", "s/ep", "total_epochs", "hrs for 300 epochs"]]
demo.sort_values("Score", ascending=True, inplace=True)
demo.style.format("{:.5f}").background_gradient(cmap="magma", vmax=demo["Score"].max(), subset="Score")
```
